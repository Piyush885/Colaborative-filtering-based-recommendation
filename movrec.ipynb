```python
df=pd.merge(df1,df2)
```

```python
df.head()
```

```python
#As this is a colaborative filtering so we donot need the generes of the movies.It only works on the userid and the product id
```

```python
df=df.drop('genres',axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
ratings_users=df.pivot_table(index=['userId'],columns=['title'],values='rating')
ratings_users.head()
```

```python
#So we have to remove those movies which is rated less that let say 10 users and fill 0 otherwise
ratings_users=ratings_users.dropna(thresh=10,axis=1).fillna(0)
```

```python
ratings_users.head()
```

```python
#Now we have to vectorize(Low rank matrix factorization) the userid and the movieid and correlate it.For this python has corr() function which basically optimize and find the best values of userid and movieid based on sse and gradient descent
```

```python
corr_matrix=ratings_users.corr(method='pearson')
```
